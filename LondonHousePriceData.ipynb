# Cleaning data

Instead of deleting all data with null values or replacing all null data with the median, I will implement a hybrid approach. This approach will use the median to fill in some missing values while deleting others.

```python
columns_fill_median = ['bathrooms', 'livingRooms', 'history_percentageChange', 'history_numericChange']

for column in columns_fill_median:
    df[column] = pd.to_numeric(df[column], errors='coerce')
    df[column].fillna(df[column].median(), inplace=True)

# Have to do this separately as 'currentEnergyRating' requires letters
df['currentEnergyRating'].fillna(df['currentEnergyRating'].mode()[0], inplace=True)
```

# Create a Correlation Matrix

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['outcode_encoded'] = le.fit_transform(df['outcode'])
```

```python
corr_variables = ["bedrooms", 'latitude', 'longitude', 'floorAreaSqM', 'saleEstimate_currentPrice', 'rentEstimate_currentPrice', 'outcode_encoded']
df_variables = df[corr_variables]

corr_matrix = df_variables.corr()

print(corr_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```
